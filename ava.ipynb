```python
import re
```

```python
with open("lexc.txt", "r", encoding="utf-8") as f:
    file = f.read()
    
print(file)
```

```python
new_file = re.sub("%", "", file)
new_file = re.sub("!", "#", new_file)
print(new_file)
```

```python
spl = new_file.split("\n")[4517:6880]
print(spl)
```

```python
res = []

for i in spl:
    if len(i.split()) > 1:
        res.append(i.split()[1])
tags = [i for i in list((set(res))) if i.startswith("N-")]
print(tags)
```

```python
patterns = []

for tag in tags:
    patterns.append(f"{tag} [<n>:] {tag}Infl (Clitics | LOC-CASES)")
    
print("\n".join(patterns))
```

```python
noun = []

for tag in tags:
    noun.append(f"LEXICON {tag}")
    for line in spl:
        if re.search(f'{tag}', line) != None:
            noun.append(re.sub(f"{tag} ; ", "", line))
    noun.append("\n")
    
print("\n".join(noun))
```

```python
line = """тӏоцебесеб:тӏоце A-TIOCEBESEB ; # "первый"
цебесеб:це A-TIOCEBESEB ; # "прежний"

хӏажатаб:хӏажата A-KUDIJA ; # "нужный, требуемый; необходимый"
бащдаб:бащда A-KUDIJA ; # "пол-, половина"
оперативияб:оперативия A-KUDIJA ; # "оперативный"
"""

adj = []

line = line.split("\n")

for each in line:
    if re.search(r'A-KUDIJA', each) != None:
        adj.append(re.sub("A-KUDIJA ; ", "", each))
        
print("\n".join(adj))
```
