```python
import re
```

1. reading copied and saved as txt lexc-compiler

```python
with open("lexc.txt", "r", encoding="utf-8") as f:
    file = f.read()
    
print(file)
```

2. removing "%" symbol and replacing symbol for comments

```python
new_file = re.sub("%", "", file)
new_file = re.sub("!", "#", new_file)
print(new_file)
```

3. analysing structure of syntax while processing numerals

```python
new_file = new_file.split("\n")
```

```python
num = []

for line in new_file:
    if "num" in line or "Num" in line or "NUM" in line:
        num.append(line)
        
print("\n".join(num))
```

4. replacing numeral tags to get list of lexicons for the optimized version

```python
for i in range(len(num)):
    if re.search("<num>", num[i]) != None:
        correct = re.sub("<num>", "", num[i])
        num[i] = f"{i}. {correct}"
    else:
        num[i] = f"{i}. {num[i]}"
        
print("\n".join(num))
```

5. dividing "LEXICON Numerals" into groups of stems according to the following tag

```python
stems_num = []

for item in num[20:58]:
    stem = item.split()[2]
    if stem not in stems_num:
        stems_num.append(stem)
        
print(stems_num)
```

6. remain only necessary numeral tags

```python
stems_num = [stem for stem in stems_num if "NUM" in stem]
print(stems_num)
```

```python
numerals = []

for stem in stems_num:
    numerals.append(f"LEXICON {stem}")
    for line in num[20:58]:
        line = " ".join(line.split()[1:])
        if re.search(f'{stem} ;', line) != None:
            numerals.append(re.sub(f"{stem} ; ", "", line))
    numerals.append("\n")
    
print("\n".join(numerals))
```

8. processing adjectives

```python
adj = []

for line in new_file:
    if "adj" in line or "Adj" in line or ((len(line.split()) > 1) and (line.split()[1].startswith("A-"))) or "ADJEC" in line:
        adj.append(line)
        
print("\n".join(adj))
```

```python
for i in range(len(adj)):
    if adj[i].startswith("<adj>"):
        correct = re.sub("<adj>", "", adj[i])
        adj[i] = f"{i}. {correct}"
    else:
        adj[i] = f"{i}. {adj[i]}"
        
print("\n".join(adj))
```

```python
stems_adj = []

for item in adj[52:]:
    stem = item.split()[2]
    if stem not in stems_adj:
        stems_adj.append(stem)
        
print(stems_adj)
```

```python
stems_adj = [stem for stem in stems_adj if "A-" in stem]
print(stems_adj)
```

```python
adjectives = []

for stem in stems_adj:
    adjectives.append(f"LEXICON {stem}")
    for line in adj[46:]:
        line = " ".join(line.split()[1:])
        if re.search(f'{stem} ;', line) != None:
            adjectives.append(re.sub(f"{stem} ; ", "", line))
    adjectives.append("\n")
    
print("\n".join(adjectives))
```

10. processing nouns

```python
noun = []

for line in new_file:
    if "<n>" in line or "Noun" in line or "N-" in line:
        noun.append(line)
        
print("\n".join(noun))
```

```python
for i in range(len(noun)):
    if noun[i].startswith("<n>"):
        correct = re.sub("<n>", "", noun[i])
        noun[i] = f"{i}. {correct}"
    else:
        noun[i] = f"{i}. {noun[i]}"
        
print("\n".join(noun))
```

```python
stems_noun = []

for item in noun[1933:4129]:
    stem = item.split()[2]
    if stem not in stems_noun:
        stems_noun.append(stem)
        
print(stems_noun)
```

```python
stems_noun = [stem for stem in stems_noun if "N-" in stem]
print(stems_noun)
```

```python
nouns = []

for stem in stems_noun:
    nouns.append(f"LEXICON {stem}-Root")
    for line in noun[1933:4129]:
        line = " ".join(line.split()[1:])
        if re.search(f'{stem} ;', line) != None:
            nouns.append(re.sub(f"{stem} ; ", "", line))
    nouns.append("\n")
    
print("\n".join(nouns))
```
